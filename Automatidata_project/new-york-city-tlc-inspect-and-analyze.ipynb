```python
# Sort the data by trip distance from maximum to minimum value
df_sorted = df.sort_values(by=['trip_distance'], ascending=False)
df_sorted[["trip_distance", "total_amount"]]
```

> Many values of trip_distance are 0, but there is still a charge applied.

* Total_amount has many negative values.

```python


# What is the average tip for trips paid for with credit card?
df_paid_with_credit = df[df["payment_type"] == 1]
print("the average tip for trips paid for with credit card: ", df_paid_with_credit["tip_amount"].mean())

# What is the average tip for trips paid for with cash?
df_paid_with_cash = df[df["payment_type"] == 2]
print("the average tip for trips paid for with cash: ", df_paid_with_cash["tip_amount"].mean())
```

```python

# How many times is each vendor ID represented in the data?
df["VendorID"].value_counts()
```
